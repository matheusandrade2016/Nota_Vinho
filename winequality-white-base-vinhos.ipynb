```r
# Carregando nossa base novamente
dados = read.csv(file = "../input/winequalitywhite/winequality-white.csv", sep = ';')
```

```r
# Verificando meus dados
str(dados)
```

```r
# Transformando nossa varival quality em fator
dados$quality = factor(dados$quality)
```

```r
str(dados)
```

```r
# Importando nossa biblioteca para teste com algoritimo Árvores de Decisão 
library(rpart)
```

```r
# Criando nossa divisão 
divisao = sample.split(dados$quality, SplitRatio = 0.75)
```

```r
# Criando minha variavel de treino
dados_treinamento = subset(dados, divisao == TRUE)
```

```r
# Criando meus dados de Teste
dados_teste = subset(dados, divisao == FALSE)
```

```r
# Criando meu modelo de arvore de decisão
modelo_arvore_de_decisao = rpart(formula = quality ~., data = dados_treinamento, control = rpart.control(minbucket = 1))
```

```r
# Verificando nosso modelo
modelo_arvore_de_decisao
```

```r
# Criando minhas previsoes para arvore de decisão
previsoes_arvore_de_decisao = predict(modelo_arvore_de_decisao, newdata = dados_teste[-12], type = 'class')
```

```r
# Verificando minhas previsões
previsoes_arvore_de_decisao
```

```r
# Verificando nossa matriz de confusão
matriz_de_confusao_arovore_de_decisao = table(dados_teste[, 12], previsoes_arvore_de_decisao)
matriz_de_confusao_arovore_de_decisao 
```

```r
# Verificando minha acuracia do modelo atravez da biblioteca 'caret'
confusionMatrix(matriz_de_confusao_arovore_de_decisao)
```

```r
# Carregando nossa base novamente
dados = read.csv(file = "../input/winequalitywhite/winequality-white.csv", sep = ';')
```

```r
# Criano minha divisao na base de dados
divisao = sample.split(dados$quality, SplitRatio = 0.75)
```

```r
# Criando minha variavel de Treinamento e teste
dados_treinamento = subset(dados, divisao == TRUE)
dados_teste = subset(dados, divisao == FALSE)
```

```r
dados_treinamento$quality = factor(dados_treinamento$quality)
str(dados_treinamento)
```

```r
# Instalando e Carregando nossa biblioteca RandomForest
# install.packages("RandomForest")
library(randomForest)
```

```r
# Criando meu modelo para nossao algoritmo RandomForest
modelo_random_forest = randomForest(x = dados_treinamento[-12], y = dados_treinamento$quality, ntree = 10)
```

```r
# Verificando nosso modelo
modelo_random_forest
```

```r
# Criando minhas variaveis de previsoes
previsoes_random_forest = predict(modelo_random_forest, newdata = dados_teste[-12])
```

```r
# Verificando minhas previsoes
previsoes_random_forest
```

```r
# Criando minha Matriz de Confusão
matriz_confusao_random_forest = table(previsoes_random_forest, dados_teste[, 12])
```

```r
# Verificando minha acuracia e minha matriz de confusão pela biblioteca "caret"
confusionMatrix(matriz_confusao_random_forest)
```

```r
# Testado nosso Algoritimo KNN, Carregando nossa biblioteca
library(class)
```

```r
# Criando minhas previsoes direto pelo algoritimo KNN
previsoes_knn = knn(train = dados_treinamento[, -12], test = dados_teste[, -12], cl = dados_treinamento[, 12], k = 5)
```

```r
# Verificando minhas previsoes
previsoes_knn
```

```r
# Criando minha matriz de confusão
matriz_confusao_knn = table(previsoes_knn, dados_teste[, 12])
matriz_confusao_knn
```

```r
# Verificando minha acurcia
confusionMatrix(matriz_confusao_knn)
```
